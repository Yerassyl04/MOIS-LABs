Logistic regression is a statistical method for predicting binary outcomes (two possible classes, like 0 and 1) based on input features. Itâ€™s widely used in classification problems, such as predicting whether a person has diabetes (yes or no), based on medical data.

# Example 1

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

df = pd.read_csv('diabetes.csv')

df.head()

```

```python
df.info()

```

```python
X = df.drop('outcome', axis=1)
y = df['outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# Example 2

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

data = pd.read_csv('breast_cancer.csv')

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
print(data['Bare Nuclei'].unique())
```

```python
data = data.replace('?', np.nan)

data['Bare Nuclei'] = pd.to_numeric(data['Bare Nuclei'])
data.info()
```

```python
data['Bare Nuclei'] = data['Bare Nuclei'].fillna(data['Bare Nuclei'].median())
```

```python
Xx = data.drop('Class', axis=1)
yy = data['Class']

X_trainn, X_testt, y_trainn, y_testt = train_test_split(Xx, yy, test_size=0.3)

modell = LogisticRegression()
modell.fit(X_trainn, y_trainn)

y_predd= modell.predict(X_testt)

print("Accuracy:", accuracy_score(y_testt, y_predd))
print("\nClassification Report:\n", classification_report(y_testt, y_predd))
print("\nConfusion Matrix:\n", confusion_matrix(y_testt, y_predd))
```
